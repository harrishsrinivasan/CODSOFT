# DATA ENCODING


```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
# Encode the categorial variable species as numeric values
data['species'] = LE.fit_transform(data['species'])
data.head()
```

```python
#optional
data['species'] = data['species'].astype(float)
```

```python
data.info()
```

# PREPROCESSING OF DATA

```python
# defining the target variable 
X = data.drop(columns= ['species'])
Y = data['species']
```

```python
# scaling the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
#splitting the dataset
from sklearn.model_selection import train_test_split as tts 
x_train, x_test, y_train, y_test = tts(scaled_X, Y, train_size = 0.70, stratify = Y)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# DECISION TREES MODEL  

```python
from sklearn.tree import DecisionTreeClassifier as DTC 
classifier = DTC(class_weight= 'balanced')
classifier = DTC()
```

```python
classifier.fit(x_train, y_train)
predicted_values = classifier.predict(x_train)
```

```python
#training set evaluation
from sklearn.metrics import classification_report
print(classification_report(y_train, predicted_values))
```

```python
#test set evaluation
predicted_values = classifier.predict(x_test)
print(classification_report(y_test, predicted_values))
```

```python
print("Accuracy: ", classifier.score(x_test, y_test) * 100)
```

the accuracy of the model using decision trees is 95.5555 percent

# LOGISTIC REGRESSION MODEL

```python
# logistic regression
from sklearn.linear_model import LogisticRegression as LR
classifier = LR(class_weight = 'balanced')
```

```python
classifier.fit(x_train, y_train)
predicted_values = classifier.predict(x_train)
predicted_probabilities = classifier.predict_proba(x_test)
```

```python
#training set evaluation
from sklearn.metrics import classification_report
k = classification_report(y_train, predicted_values)
print(k)
```

```python
#test set evaluation
predicted_values = classifier.predict(x_test)
print(classification_report(y_test, predicted_values))
```

```python
print("Accuracy: ", classifier.score(x_test, y_test) * 100)
```

the accuracy of the model using logistic regression is 93.3333 percent
